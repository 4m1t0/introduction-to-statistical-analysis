##### caption 2.4
$g(c) = \sum_{i=1}^{n} \left| x_{i}-c \right|$ を最小にする $c$ が median であることを示す。  

元の数列を昇順にソートしたものを $x_{(i)} (i=1, 2, \cdots, n) (x_{1} \leq x_{2} \leq \cdots \leq x_{n})$ とする。  
$x_{(k)} \leq c \leq x_{(k+1)}$ とすると、  
$$
    \begin{align*}
        g(c) &= \sum_{i=1}^{n} \left| x_{(i)}-c \right| \\
              &= \sum_{i=1}^{k} (c - x_{(i)}) + \sum_{i=k+1}^{n} (x_{(i)}-c) \\
              &= kc - \sum_{i=1}^{k}x_{(i)} + \sum_{i=k+1}^{n}x_{(i)} + (n-k)c \\
              &= c(2k-n) - \sum_{i=1}^{k}x_{(i)} + \sum_{i=k+1}^{n}x_{(i)}
    \end{align*}
$$

上式を $n$ が奇数と偶数である場合に分けて考える。  

$n=2m+1$ のとき、$0 \leq k \leq m$ は $x_{(k)} \leq c \leq x_{(k+1)}$ より $c \leq x_{(m)} \leq x_{(m+1)}$ であり、このとき $g(c)$ は単調減少の関数となる。  
一方で　$m+1 \leq k \leq n$ の場合は $x_{(m+1)} \leq c$ でもあり、このとき $g(c)$ は単調増加の関数となる。  
よって、$n$ が奇数のときは $b=x_{(m+1)}$ のときに最小になり、これは median が奇数のときの定義に一致する。

$n=2m$ のとき、$0 \leq k \leq m-1$ は $x_{(k)} \leq c \leq x_{(k+1)}$ より $c \leq x_{(m)}$ であり、このとき $g(c)$ は単調減少の関数となる。  
$k=m$ のときは $x_{(m)} \leq c \leq x_{(m+1)}$ であり、このとき $g(c)=-\sum_{i=1}^{m}x_{(i)}+\sum_{i=m+1}^{n}x_{(i)}$ となり、$g(c)$ は一定の値を取る。  
$m+1 \leq k \leq n$ は $x_{(m+1)} \leq c$ であり、このとき $g(c)$ は単調増加の関数となる。  
よって、$n$ が偶数のときは $c$ は $x_{(m)} \leq c \leq x_{(m+1)}$ のときに $g(c)$ は最小になり、その値は $\frac{x_{(m)} + x_{(m+1)}}{2}$ である。  
これは median が偶数のときの定義に一致する。

元の数列を昇順にソートしたものを $x_{(i)} (i=1, 2, \cdots, n) (x_{1} \leq x_{2} \leq \cdots \leq x_{n})$ とする。  
平均の定義より、$n \bar{x} = \sum_{i=1}^{n}x_(i)$ である。  
$\bar{x} \lt x_{(1)}$ の場合、$n x_{(1)} \lt \sum_{i=1}^{n}x_{(i)}$ であるため、$x_{(1)} \leq \bar{x}$ となる。  
同様にして $x_{(n)} \lt \bar{x}$ の場合、$\sum_{i=1}^{n}x_{(i)} \lt n x_{(n)}$ であるため、$\bar{x} \leq x_{(n)}$ となる。  
以上より、$x_{(1)} \leq \bar{x} \leq x_{(n)}$ である。

#### exercises 2.2.7.2
$n$ に関する数学的帰納法により証明する。

$n=1$ のとき、算術平均と幾何平均の値は一致する。

$n=k$ のとき、定理の主張 $(x_1x_2\cdots x_n)^{\frac{1}{n}} \leq \bar{x}$ が成立すると仮定し、$n=k+1$ の場合を考える。  
$$
    \begin{align*}
        \frac{x_1+x_2+ \cdots + x_k + x_{k+1}}{k+1} - (x_1x_2\cdots x_k x_{k+1})^{\frac{1}{k+1}} \geq 0 \label{1}\tag{1}
    \end{align*}
$$
が成立することを示せば良い。

$(\ref{1})$ を変形した
$$
    \begin{align*}
        f(x_{k+1}) = \left(\frac{x_1+x_2+\cdots+x_k+x_{k+1}}{k+1}\right)^{k+1}-x_1x_2 \cdots x_kx_{k+1}
    \end{align*}
$$
について、$f(x_{k+1}) \geq 0$ であることを証明する。

$x_{k+1}=c$ とし、$f(c)$ を $c$ について微分すると、
$$
    \begin{align*}
        f'(c) &= (k+1)\left(\frac{x_1+x_2+\cdots+x_k+c}{k+1}\right)^{k}\left(\frac{x_1+x_2+\cdots+x_k+c}{k+1}\right)'-x_1x_2\cdots x_k \\
            &= (k+1)\left(\frac{x_1+x_2+\cdots+x_k+c}{k+1}\right)^{k}\frac{1}{k+1}-x_1x_2\cdots x_k \\
            &= \left(\frac{x_1+x_2+\cdots+x_k+c}{k+1}\right)^{k} - x_1x_2\cdots x_k.
    \end{align*}
$$
さらに微分すると、
$$
    \begin{align*}
        f''(c) &= k\left(\frac{x_1+x_2+\cdots+x_k+c}{k+1}\right)^{k-1}\frac{1}{k+1} \\
            &= \frac{k}{k+1}\left(\frac{x_1+x_2+\cdots+x_k+c}{k+1}\right)^{k-1}.
    \end{align*}
$$
$k=0$ のとき $f''(c) = 0$ であり、$k=1$ のとき $f''(c)=\frac{1}{2}\gt0$ である。  
また、$k\gt1$ のとき $c>0$ ならば、$x_1+x_2+\cdots+x_k+c \gt 0$なので、$f''(c) \gt 0.$  
以上より $f''(c)$ が区間 $[0, \infty)$ において下に凸な関数であることから $f'(c)$ は区間 $[0, \infty)$ において単調増加である。

$\bar{x}=\frac{x_1+x_2+\cdots+x_k}{k}$ とすると、帰納法の仮定より $\bar{x} \geq (x_1x_2\cdots x_k)^{\frac{1}{k}}.$  
また、
$$
    \begin{align*}
        \frac{x_1+x_2+\cdots+x_k+\bar{x}}{k+1} = \frac{k\bar{x}+\bar{x}}{k+1} = \bar{x}. \label{2}\tag{2}
    \end{align*}
$$
よって、$f'(\bar{x})=\bar{x}^{k}-x_1x_2\cdots x_k \geq 0.$  
$\bar{x}\gt0$ であり、区間 $[0, \infty)$ において $f'(c)$ は単調増加なので、区間 $[\bar{x}, \infty)$ において $f'(c)\gt0$ が成り立つ。  
さらに、$f(\bar{x})$ に対して $(\ref{2})$ を用いると
$$
    \begin{align*}
        f(\bar{x}) &= \left(\frac{x_1+x_2+\cdots+x_k+\bar{x}}{k+1}\right)^{k+1}-x_1x_2 \cdots x_k\bar{x} \\
            &= \bar{x}^{k+1} -x_1x_2 \cdots x_k\bar{x} \\
            &= \bar{x}(\bar{x}^{k}-x_1x_2 \cdots x_k) \\
            &= \bar{x}\lbrace\left(\frac{x_1+x_2+\cdots+x_k+\bar{x}}{k+1}\right)^{k}-x_1x_2 \cdots x_k\rbrace \\
            &= \bar{x}f'(\bar{x}) \geq 0.
    \end{align*}
$$
したがって、$c\geq \bar{x}$ となるすべての実数 $c$ に対して $f(c)\geq0$ が成り立つ。  
$x_1, x_2, \cdots, x_k$ を昇順に並べたものを$x_{(1)}, x_{(2)}, \cdots, x_{(k)}$ とすると、
$$
    \begin{align*}
        x_{(k+1)} = \frac{nx_{(k+1)}}{n} \geq \frac{x_{(1)}+x_{(2)}+\cdots+x_{(k)}}{k} = \bar{x}.
    \end{align*}
$$
よって、$f(x_{(k+1)})\geq0.$  
$f(x_{k+1})$ の値は順序に依存しないため、$f(x_{k+1})\geq0$ であり、これより $n=k+1$ のときの不等式 $(x_1x_2\cdots x_n)^{\frac{1}{n}} \leq \bar{x}$ が得られる。

以上より、全ての正の数 $n$ について $(x_1x_2\cdots x_n)^{\frac{1}{n}} \leq \bar{x}$ が成立することが証明された。

#### exercises 2.2.7.3
$(x_1x_2\cdots x_n)^{\frac{1}{n}} \leq \frac{1}{n}\sum_{i=1}^{n}x_i$ の対数を取ると、
$$
    \begin{align*}
        \log{(x_1x_2\cdots x_n)^{\frac{1}{n}}} \leq \log{\left(\frac{1}{n}\sum_{i=1}^{n}x_i\right)}
    \end{align*}
$$
左辺を変形すると、
$$
    \begin{align*}
        \log{(x_1x_2\cdots x_n)^{\frac{1}{n}}} &= \frac{1}{n}(\log{x_1}+\log{x_2}+\cdots+\log{x_n}) \\
            &= \frac{1}{n}\sum_{i=1}^{n}\log{x_i}
    \end{align*}
$$
よって、$\frac{1}{n}\sum_{i=1}^{n}\log{x_i} \leq \log{\left(\frac{1}{n}\sum_{i=1}^{n}x_i\right)}.$
